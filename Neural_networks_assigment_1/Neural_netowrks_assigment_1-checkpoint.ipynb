```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np
import statistics as st
```

```python
dataset = pd.read_csv('NeuralDataset.csv')
print(dataset.shape)

dft1 = dataset['Feature1']
dft2 = dataset['Feature2']
cstrs = dataset['Clusters']

print(dataset.head())
filtered_label1 = dataset[dataset.Clusters == 0]
filtered_label2 = dataset[dataset.Clusters == 1]

feature1ofcluster1 = filtered_label1['Feature1']
feature2ofcluster1 = filtered_label1['Feature2']
print(feature1ofcluster1)
print(feature2ofcluster1)



feature1ofcluster2 = filtered_label2['Feature1']
feature2ofcluster2 = filtered_label2['Feature2']
print(feature1ofcluster1)
print(feature2ofcluster1)
```

```python
plt.scatter(feature1ofcluster1,feature2ofcluster1,c='green')
plt.scatter(feature1ofcluster2,feature2ofcluster2,c='red')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Clusters')

plt.text(0.25, 0.33, 'Cluster No.1', fontsize=10)
plt.text(0.7, 0.16, 'Cluster No.2', fontsize=10)
plt.grid()
plt.show();
```

```python
import neurolab as nl
import matplotlib.pyplot as plt
input = [dft1,dft2]
target = [[0],[1]]
# Create net with 2 inputs and 1 neuron
net = nl.net.newp(nl.tool.minmax(input), 1)
# train with delta rule (default)
error = net.train(input, target,epochs=100, show=1, lr=0.3)
# Plot results
plt.plot(error)
plt.xlabel('Epoch number')
plt.ylabel('Train error')
plt.grid()
plt.show();
```
