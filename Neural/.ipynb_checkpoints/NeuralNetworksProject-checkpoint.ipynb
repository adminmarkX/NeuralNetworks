```python
# Importing Libraries
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_excel('fertility.xlsx')
dataset
```

<p> We need to represend data in form of numbers so the whole dataset to be classified properly <p>

```python
dataset['Diagnosis'] = dataset['Diagnosis'].replace(['N'],0)     
dataset
```

```python
dataset['Diagnosis'] = dataset['Diagnosis'].replace(['O'],1)     
dataset
```

<p> We are sorting the dataset in order of Diagnosis columns which created immidietly the Target dataset for the neural network training ( Activation function split )<p>

```python
datasetNonDublicates = dataset.drop_duplicates(keep='first')
datasetNonDublicates.head(50)
```

```python
datasetSorted = datasetNonDublicates.sort_values(by='last')
datasetSorted.head(50)
```

```python
for x in ['Age']:
    q75,q25 = np.percentile(dataset.loc[:,x],[75,25])
    intr_qr = q75-q25
 
    max = q75+(1.5*intr_qr)
    min = q25-(1.5*intr_qr)
 
    dataset.loc[dataset[x] < min,x] = np.nan
    dataset.loc[dataset[x] > max,x] = np.nan
```

```python
import neurolab as nl

np.random.seed(42)
net = nl.net.newff(nl.tool.minmax(X), [10,2]) #2 because we use 2 bits for each target
#change the activation function for the output layer
net.layers[0].transf = nl.trans.TanSig() #hidden layer
net.layers[-1].transf = nl.trans.TanSig() #hidden layer
net
```

```python
np.random.seed(42)
# train the network
error = net.trainf = nl.train.train_gda(net,X_train, y_train, epochs=1000, show=100, lr =0.3 ,goal=0.0005)

# Plot results
plt.plot(error)
plt.xlabel('Epoch number')
plt.ylabel('Train error')
plt.grid()
plt.show();
```
