```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
import numpy as np
import statistics as st
import neurolab as nl
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
dataset = pd.read_csv('wine.csv')
dataset.shape
```

```python
dataset.head(500)
```

```python
target = dataset['Category']
target.head(500)
```

```python
dataset.pop('Category')
```

```python
dataset.head(500)
```

<h3> Silhouette analysis <h3>

```python
X = dataset
y = target
```

```python
np.random.seed(42)
sse_ = []
for k in range(2, 100):
 kmeans = KMeans(n_clusters=k).fit(X)
 sse_.append([k, silhouette_score(X, kmeans.labels_)])
```

```python
plt.plot(pd.DataFrame(sse_)[0], pd.DataFrame(sse_)[1]);
```

```python
from yellowbrick.cluster import KElbowVisualizer
```

```python
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10), timings=False ,locate_elbow=False,random_state=42)
visualizer.fit(X) # Fit the data to the visualizer
visualizer.show() # Finalize and render the figure
```

```python
import neurolab as nl
import pylab as pl
```

```python
# Scale the data [0, 1]
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
X = min_max_scaler.fit_transform(X)
```

```python
np.random.seed(42)
net = nl.net.newc(nl.tool.minmax(X), 3)

# we choose 3 because we know that there are 3 categories
error = net.trainf = nl.train.train_wta(net, X, epochs = 1000,show = 200)

# Plot results:
plt.title('Example 3')
plt.subplot(2,1,1)
plt.plot(error)
plt.xlabel('Epoch number')
plt.ylabel('error (default MAE)')
w = net.layers[0].np['w']
plt.subplot(2,1,2)
plt.plot(X[:,2], X[:,3], '.', \
w[:,2], w[:,3], 'p')
plt.legend(['train samples', 'train centers'])
plt.show();
```
